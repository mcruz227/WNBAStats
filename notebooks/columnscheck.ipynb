```python
import pandas as pd

file_path = ("../backend/data/wnba_cleaned.csv")
df = pd.read_csv(file_path)

print("Columns in dataset:")
print(df.columns)
```

```python
print(df.isnull().sum())




```

```python
print("Duplicate rows:", df.duplicated().sum())

```

```python
print(df.dtypes)
df['game_date'] = pd.to_datetime(df['game_date'])
df['game_date_time'] = pd.to_datetime(df['game_date_time'])
df['team_uid'] = pd.to_numeric(df['team_uid'], errors='ignore') 
print(df['team_home_away'].unique())  #shows me if theres anything that is not home or away
print(df['team_winner'].unique())  #shows me if something not false or true



```

```python
df.to_csv("../backend/data/wnba_cleaned.csv", index=False)
```
